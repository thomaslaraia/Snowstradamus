```python
from matplotlib import pyplot as plt
import numpy as np

x_values = []
y_values = []

for i in range(2, 101):
    for j in range(1,i):
        x_values.append(j / i)
        y_values.append(np.random.random(1)/(j))

plt.plot(x_values, y_values, 'bo', markersize=.5)  # 'bo' specifies blue dots
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_values = []
y_values = []
A = []
for _ in range(10000):
    i = np.random.poisson(35)  # Poisson distribution with lambda=35, can reasonably reach 60 (total photons, denominator)
    A.append(i)
    j = np.random.randint(1, i + 1)  # Random integer from 1 to i (numerator)
    x_values.append(j / i)
    y_values.append(np.random.random(1) / j)

# print(max(A))
plt.plot(x_values, y_values, 'bo', markersize=.5)  # 'bo' specifies blue dots
plt.show()
```
