```python
from scripts.imports import *

df = pd.read_pickle('dataset_lcforest_outlier22_th3_1km_noprior_ta.pkl')

df_willow = df[(df['camera']=='willowcreek')&(df['file_index']==3)].copy()
```

```python
from scripts.imports import *
import rioxarray

filepath = '../scratch/data/DW/DynamicWorld_willowcreek_20240721.tif'

# Read the file as an xarray.DataArray with lat/lon coordinates
da = rioxarray.open_rasterio(filepath, masked=True).rio.reproject("EPSG:4326")

da

# # Select the first band
# band1 = da.sel(band=1)

# # Mask out values that are not 1, 5, or 6
# masked = band1.where(band1.isin([0, 5, 6]))

# # Plot the masked data
# masked.plot.imshow()  # Adjust colormap if needed

# plt.title("Dynamic World - Willow Creek (Classes 1, 5, 6)")
# plt.xlabel("Longitude")
# plt.ylabel("Latitude")
# plt.show()
```

```python
# Interpolate values at the given lat/lon points using xarray
df_willow["land_cover_label"] = da.sel(band=1).interp(
    y=("points", df_willow.latitude.values),
    x=("points", df_willow.longitude.values),
    method="nearest"
).values
```
