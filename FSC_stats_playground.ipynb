# Performance vs msw_flag

```python
from scripts.hist_plot import *
from scripts.confusion import *

# Load the dataframe
df = pd.read_pickle('five_sites_0-05_0-005box_snowreffed.pkl')

hue_labels = {'0.0': 'No Snow', '1.0': 'Ground Snow', '2.0': 'Ground+Canopy Snow'}

accuracy = []
msw_flag_max = []
size = []

for i in np.arange(5,0,-.25):
    msw_flag_max.append(i)

    loc_df = df[(df['msw'] < i)&(df['Confidence'] == 1)].reset_index(drop=True)
    size.append(loc_df.shape[0])

    FI = hist_plot(loc_df, hue_labels, X='asr', Hue='JointSnow', plot=False)

    acc = confusion(loc_df, FI,plot=False)
    
    accuracy.append(acc)
```

```python
plt.plot(msw_flag_max,accuracy)
plt.title('NoSnow/Snow accuracy')
plt.xlabel('maximum msw_flag')
plt.ylabel('Accuracy')
plt.savefig('./images/500mres/conf2_msw')
# plt.plot(msw_flag_max,np.array(size)/2000)
```
