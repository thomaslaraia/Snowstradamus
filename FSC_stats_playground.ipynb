# Scatter Plots

```python
loc_df = df[(df['msw'] < 2.1)&(df['Confidence'] == 1)].reset_index(drop=True)

plt.figure()

hue_labels = {'0.0': 'No Snow', '1.0': 'Ground Snow', '2.0': 'Ground+Canopy Snow'}

scatter = sns.scatterplot(loc_df, x='MOD10A1F', y = 'asr' ,hue='JointSnow', palette='tab10', style='camera',size='msw',
               legend='brief')
handles, labels = scatter.get_legend_handles_labels()
new_labels = [hue_labels[label] if label in hue_labels else label for label in labels]

scatter.legend(handles=handles, labels=new_labels, loc='upper right')

# plt.plot([.188,.188],[0,5], 'r--')
# plt.plot([.462,.462],[0,5], 'r--')
# plt.legend(loc='upper right')
# plt.xlim(0,10)
# plt.ylim(0,9)
# plt.savefig('./images/500mres/scatter_msw2_Eg')
plt.show()
```
