```python
from scripts.imports import *
import xarray as xr
import rasterio
from scipy.spatial import cKDTree
from scipy.spatial.distance import pdist
from pyproj import Transformer, CRS
from rasterio.windows import from_bounds as windows_from_bounds
import matplotlib.pyplot as plt

def calculate_bounds(center_lat, center_lon, radius):
    radius_deg_lat = radius / 111000
    radius_deg_lon = radius / (111000 * np.cos(np.radians(center_lat)))
    lat_min = center_lat - radius_deg_lat
    lat_max = center_lat + radius_deg_lat
    lon_min = center_lon - radius_deg_lon
    lon_max = center_lon + radius_deg_lon
    return lat_min, lat_max, lon_min, lon_max

def calculate_dissimilarity(data):
    valid_pixels = data[~np.isnan(data)]
    if len(valid_pixels) > 1:
        return np.mean(pdist(valid_pixels[:, None]))
    return np.nan

def apply_elevation_mask(cropped_reprojected_data, elevation_data, tolerance, location_elevation, percentage=0.80):
    mask = np.zeros_like(cropped_reprojected_data, dtype=bool)
    pixel_size_y = elevation_data.shape[0] // cropped_reprojected_data.shape[0]
    pixel_size_x = elevation_data.shape[1] // cropped_reprojected_data.shape[1]

    valid_pixels_total = 0
    pixels_meeting_percentage = 0
    elevation_block_nan_count = 0

    for i in range(cropped_reprojected_data.shape[0]):
        for j in range(cropped_reprojected_data.shape[1]):
            elev_block = elevation_data[i*pixel_size_y:(i+1)*pixel_size_y, j*pixel_size_x:(j+1)*pixel_size_x]
            elev_values = elev_block[~np.isnan(elev_block)]

            if elev_values.size == 0:
                elevation_block_nan_count += 1
                continue

            within_tol = (elev_values >= location_elevation - tolerance) & (elev_values <= location_elevation + tolerance)
            frac_within = np.sum(within_tol) / elev_values.size

            if not np.isnan(cropped_reprojected_data[i, j]):
                valid_pixels_total += 1
                if frac_within >= percentage:
                    pixels_meeting_percentage += 1
                    mask[i, j] = True

    masked_snow_cover_data = np.where(mask, cropped_reprojected_data, np.nan)

    print(
        f"  ElevMask: tol=±{tolerance}m | elevation blocks skipped (NaN): {elevation_block_nan_count} | "
        f"pixels with snow: {valid_pixels_total} | passing elevation threshold: {pixels_meeting_percentage}"
    )

    return masked_snow_cover_data


def perform_analysis(center_lat, center_lon, radius, location_elevation, zarr_path, dem_path, tolerances, percentage=0.80):
    lat_min, lat_max, lon_min, lon_max = calculate_bounds(center_lat, center_lon, radius)

    # === Load SCFG Zarr ===
    ds = xr.open_zarr(zarr_path)
    scfg = ds['scfg'].sel(lat=slice(lat_min, lat_max), lon=slice(lon_min, lon_max))

    # === Load DEM & subset ===
    with rasterio.open(dem_path) as src:
        wgs84 = CRS.from_epsg(4326)
        tif_crs = src.crs
        transformer = Transformer.from_crs(wgs84, tif_crs, always_xy=True)
        center_x, center_y = transformer.transform(center_lon, center_lat)
        min_x = center_x - radius
        max_x = center_x + radius
        min_y = center_y - radius
        max_y = center_y + radius
        window = windows_from_bounds(min_x, min_y, max_x, max_y, src.transform)
        elevation_data = src.read(1, window=window)
        elevation_data = np.where(elevation_data == src.nodata, np.nan, elevation_data)

    # === Reference pixel ===
    scfg_lat = scfg.lat.values
    scfg_lon = scfg.lon.values
    scfg_shape = (len(scfg_lat), len(scfg_lon))
    ref_lat_idx = np.argmin(np.abs(scfg_lat - center_lat))
    ref_lon_idx = np.argmin(np.abs(scfg_lon - center_lon))

    monthly_dissimilarity_results = {m: np.zeros(len(tolerances)) for m in range(1, 13)}
    monthly_unmasked_dissimilarity = {m: 0 for m in range(1, 13)}
    valid_days_count = {m: np.zeros(len(tolerances)) for m in range(1, 13)}

    for t in scfg.time.values:
        t_pd = pd.to_datetime(t)
        month = t_pd.month
        try:
            scfg_vals = scfg.sel(time=t).values
        except Exception:
            continue

        if np.isnan(scfg_vals).all():
            continue

        ref_val = scfg_vals[ref_lat_idx, ref_lon_idx]
        if np.isnan(ref_val):
            continue

        delta_vals = scfg_vals - ref_val

        # Unmasked dissimilarity
        mean_unmasked = calculate_dissimilarity(delta_vals)
        if not np.isnan(mean_unmasked):
            monthly_unmasked_dissimilarity[month] += mean_unmasked

        # Masked dissimilarity
        for k, tol in enumerate(tolerances):
            masked_vals = apply_elevation_mask(delta_vals, elevation_data, tol, location_elevation, percentage)
            valid_count = np.sum(~np.isnan(masked_vals))
            if valid_count > 1:
                mean_dissimilarity = calculate_dissimilarity(masked_vals)
                monthly_dissimilarity_results[month][k] += mean_dissimilarity
                valid_days_count[month][k] += 1
            # Optional debug:
            # else:
            #     print(f"Month {month}, tol {tol} — no valid pixels.")

    # Normalize
    for m in range(1, 13):
        with np.errstate(divide='ignore', invalid='ignore'):
            monthly_dissimilarity_results[m] /= np.where(valid_days_count[m] == 0, 1, valid_days_count[m])
            monthly_unmasked_dissimilarity[m] /= np.max(valid_days_count[m]) if np.max(valid_days_count[m]) > 0 else 1

    return monthly_dissimilarity_results, monthly_unmasked_dissimilarity

# === Parameters ===
center_lat = 45.8238
center_lon = 7.5609
radius = 5000
location_elevation = 2091
zarr_path = '../scratch/data/SCFG_zarrs/torgnon.zarr'
dem_path = '../scratch/data/landsat_masking/ALOS_AW3D30_Elevation_torgnon.tif'
tolerances = np.arange(50, 251, 10)

monthly_dissimilarity_results, monthly_unmasked_dissimilarity = perform_analysis(
    center_lat, center_lon, radius, location_elevation, zarr_path, dem_path, tolerances, percentage=0.80
)

# === Plotting ===
plt.figure(figsize=(12, 8))
colors = plt.cm.coolwarm(np.linspace(0, 1, 12))

for month in range(1, 13):
    plt.plot(tolerances, monthly_dissimilarity_results[month], marker='o', color=colors[month-1], label=f'Month {month} (Masked)')
    plt.plot([tolerances.min(), tolerances.max()], [monthly_unmasked_dissimilarity[month]]*2, color=colors[month-1], linestyle='--', label=f'Month {month} (Unmasked)')

plt.xlabel('Elevation Tolerance (m)')
plt.ylabel('Average Within-Cluster Dissimilarity')
plt.title('SCFG Dissimilarity vs Elevation Threshold (Torgnon)')
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import xarray as xr
import numpy as np
import pandas as pd
import rioxarray
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# === Load SCFG example day ===
scfg_file = '../scratch/data/SCFG_zarrs/torgnon.zarr'
ds_scfg = xr.open_zarr(scfg_file)
scfg_day = ds_scfg.sel(time="2021-01-15")['scfg']  # pick a day with data

# === Load DEM ===
dem = rioxarray.open_rasterio('../scratch/data/landsat_masking/ALOS_AW3D30_Elevation_torgnon.tif').squeeze()
dem_data = dem.values

# === Match spatial dimensions ===
scfg_data = scfg_day.values
scfg_shape = scfg_data.shape
dem_shape = dem_data.shape
block_size_y = dem_shape[0] // scfg_shape[0]
block_size_x = dem_shape[1] // scfg_shape[1]

print(f"SCFG shape: {scfg_shape}, DEM shape: {dem_shape}")
print(f"Block size per SCFG pixel: {block_size_y} × {block_size_x}")

# === Collect elevation stats per SCFG pixel ===
records = []
for i in range(scfg_shape[0]):
    for j in range(scfg_shape[1]):
        dem_block = dem_data[i*block_size_y:(i+1)*block_size_y, j*block_size_x:(j+1)*block_size_x]
        valid = dem_block[~np.isnan(dem_block)]
        if valid.size == 0:
            continue
        records.append({
            "i": i,
            "j": j,
            "mean": np.mean(valid),
            "std": np.std(valid),
            "min": np.min(valid),
            "max": np.max(valid),
            "range": np.max(valid) - np.min(valid),
            "count": valid.size
        })

elev_stats_df = pd.DataFrame(records)
print(elev_stats_df.head(20))

# === Plot the DEM and SCFG pixel grid ===
fig, ax = plt.subplots(figsize=(10, 10))
dem_plot = ax.imshow(dem_data, cmap='gray', origin='upper')
plt.colorbar(dem_plot, ax=ax, label='Elevation (m)')

# Overlay SCFG grid
for _, row in elev_stats_df.iterrows():
    i = int(row['i'])
    j = int(row['j'])
    mean_elev = row['mean']
    rect = Rectangle(
        (j * block_size_x, i * block_size_y),
        block_size_x, block_size_y,
        linewidth=0.5,
        edgecolor='red',
        facecolor='none'
    )
    ax.add_patch(rect)
    # Optional: show elevation in center
    ax.text(
        j * block_size_x + block_size_x / 2,
        i * block_size_y + block_size_y / 2,
        f"{mean_elev:.0f}",
        color='blue',
        fontsize=6,
        ha='center',
        va='center'
    )

ax.set_title("DEM with SCFG Pixel Grid Overlay")
ax.set_xlabel("X (DEM column index)")
ax.set_ylabel("Y (DEM row index)")
plt.tight_layout()
plt.show()
```
